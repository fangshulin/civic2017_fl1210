```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
import statsmodels.formula.api as smf
import geopandas as gpd
```

```python
df_all_2 = pd.read_csv('df_all.csv')
```

```python
df_all_2.columns
```

```python
nyshp = gpd.read_file('geo_export_2d9f7392-d758-4afa-895d-0db3433c0083.shp')
```

```python
nyshp.head()
```

```python
nyshp['puma'] = nyshp['puma'].astype(int)
nyshp1 = nyshp.merge(df_all_2,left_on='puma',right_on='PUMA')
```

```python
fig = plt.figure(figsize = (20,5))
ax1 = fig.add_subplot(131)
nyshp1.plot(ax=ax1, color='black', linewidth=1, alpha=0.1)
nyshp1.plot(ax=ax1, column='entropy_index_9', cmap='Reds', edgecolor='black',vmin=0.78, vmax=0.98, legend=True);
plt.title('Income Integration by PUMA, 1990', size=12)
plt.grid(linewidth=0.25)


ax2 = fig.add_subplot(132)
nyshp1.plot(ax=ax2, color='black', linewidth=1, alpha=0.1)
nyshp1.plot(ax=ax2, column='entropy_index_0', cmap='Reds', edgecolor='black',vmin=0.78, vmax=0.98, legend=True);
plt.title('Income Integration by PUMA, 2000', size=12)
plt.grid(linewidth=0.25)

ax3 = fig.add_subplot(133)
nyshp1.plot(ax=ax3, color='black', linewidth=1, alpha=0.1)
nyshp1.plot(ax=ax3, column='entropy_index_1', cmap='Reds', edgecolor='black',vmin=0.78, vmax=0.98, legend=True);
plt.title('Income Integration by PUMA, 2010', size=12)
plt.grid(linewidth=0.25)

plt.savefig('Income Integration')
```
